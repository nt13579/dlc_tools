              'Specificity, True Negative Rate (TNR)',
              'Precision, Positive Predictive Value (PPV)',
              'Negative Predictive Value (NPV)',
              'False Negative Rate (FNR)',
              'False Postive Rate (FPR)',
              'False Discovery Rate (FDR)',
              'False Omission Rate (FOR)',
              'Threat Scor (TS)',
              'Accuracy (ACC)',
              'F1 Score']

for i, s in enumerate(snapshots):
    errors = s['Test'].iloc[:,::2].astype('float64')
    TP = errors['TP']
    TN = errors['TN']
    FP = errors['FP']
    FN = errors['FN']
    
    #TPR, TNR, PPV, NPV, FNR, FPR, FDR, FOR, TS, ACC, F1 = np.full(11, math.nan)

    TPR = TP/(TP+FN) #Sensitivity, True Postive Rate
    TNR = TN/(TN+FP) #Specificity, True Negative Rate
    PPV = TP/(TP+FP) #Precision, Positive Predictive Value
    NPV = TN/(TN+FN) #Negative Predictive Value
    FNR = FN/(FN+TP) #False Negative Rate
    FPR = FP/(FP+TN) #False Postive Rate
    FDR = FP/(FP+TP) #False Discovery Rate
    FOR = FN/(FN+TN) #False Omission Rate
    TS  = TP/(TP+FN+FP)           #Threat Score
    ACC = (TP+TN)/(TP+TN+FP+FN)   #Accuracy
    F1 = 2*(PPV*TPR)/(PPV+TPR)    #F1 Score

    j = len(snapshots)
    
    a = pd.concat([TPR, TNR, PPV, NPV, FNR, FPR, FDR, FOR, TS, ACC, F1], axis = 1)
    a = np.round(a, decimals = 4)
    stats.iloc[:,i::j] = a.to_numpy()
stats.loc['AVG'] = stats.mean()
